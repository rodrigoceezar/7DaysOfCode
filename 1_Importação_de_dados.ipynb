### Analisando os dados nulos

```python
cols = ['tipo_vinculo_usuario','id_emprestimo']
df_aux.loc[:,cols].groupby('tipo_vinculo_usuario').count()
```


Conclusão: O valores nulos sempre requerem uma certa atenção quanto ao seu tratamento, eles podem ser tratados da seguinte forma: 1. Substituindo por valores especificos; 2. Usando a média, mediana ou a moda; 3. Removendo os valores.

- A metodologia utilizada para o tratamento dos dados nulos da presente análise será:
1. Coluna data_renovação: Serão substituído pelo numeral 0, visto que no entendimento dos dados apresentados, quando não se tem dados da data de renovação, isso significa que o livro/exemplar não foi renovado.
2. Coluna data_devolução: Serão também substituídos por 0, tendo como entendimento que as linhas onde esteja nula são os livros que não foram devolvidos.
3. Coluna matricula_ou_siape: Após análise acima, verificou-se que os dados onde matricula_ou_siape esteja nulo podem ser os dados de usuários externos e outros, totalizando **2.648** emprestimos por pessoas sem número de matrícula, ou seja, sem vínculo. A priori, serão removidos esses dados, em outro ciclo poderá ser incluído na análise.

OBS: Após se aprofundar mais na análise, poderão ser feitas outras modificações.

### Tratando os dados nulos

```python
# Preencher os dados nulos da coluna data_renovação por 0
df1['data_renovacao'] = df1['data_renovacao'].fillna(0)

# Preencher os dados nulos da coluna data_devolução por 0
df1['data_devolucao'] = df1['data_devolucao'].fillna(0)

# Remover os dados nulos da coluna matricula_ou_siape
df1.dropna(subset=['matricula_ou_siape'], inplace=True)
```

### Perguntas #1 Days Of Code

#### A quantidade de empréstimos está aumentando ou diminuindo ao decorrer dos últimos anos?

```python
df1.head(2)
```

```python
df_aux = df1.copy()

df_aux['ano'] = df_aux.data_emprestimo.dt.year
cols = ['ano', 'id_emprestimo']

graph1 = df_aux.loc[:,cols].groupby(['ano']).count().sort_values(by='ano', ascending=False).reset_index()

plt.figure(figsize=(20,8))
plt.subplot( 1, 2, 1 )
sns.barplot(x='ano', y='id_emprestimo', data=graph1)

plt.subplot( 1, 2, 2 )
sns.regplot(x='ano', y='id_emprestimo', data=graph1, scatter=False)
plt.show()
```

Conclusão: Nota-se que ao decorrer dos anos o número de emprestimos diminuiu. Observa-se uma grande diminuição em 2020, mas isso se dá pelo fato de termos apenas os dados do primeiro semestre de 2020. O que nos faz pensar, será que há uma diferença tão grande de emprestimos no 1º e no 2º semestre?

#### Em quais bibliotecas do sistema estão a maior quantidade de empréstimos?

```python
cols = ['id_emprestimo', 'biblioteca']

graph2 = df1.loc[:,cols].groupby('biblioteca').count().sort_values('id_emprestimo', ascending=False).reset_index().head(10)

ax = sns.barplot(x='biblioteca', y='id_emprestimo', data=graph2)

# Tamanho do plot
plt.figure(figsize=(10,20))

# Formatando os rótulos dos eixos y para o formato brasileiro
formatter = ticker.StrMethodFormatter("{x:,.2f}")
ax.yaxis.set_major_formatter(formatter)

# Definindo os rótulos do eixo x de forma horizontal
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

plt.show()
```

Conclusão: A **Biblioteca Central Zila Madane** é onde possui o maior número de emprestimos de toda a comunidade universitária da UFRN, nos últimos 10 anos.

#### Quais são os temas mais emprestados? E os menos?

```python
cols = ['colecao','id_emprestimo']

graph3 = df1.loc[:,cols].groupby('colecao').count().sort_values('id_emprestimo', ascending=False).reset_index().head()
graph4 = df1.loc[:,cols].groupby('colecao').count().sort_values('id_emprestimo', ascending=True).reset_index().head()

print(f'O tema mais emprestado é {graph3.iloc[0,0]}, com {graph3.iloc[0,1]} emprestimos\nE o tema menos emprestado é {graph4.iloc[0,0]} com {graph4.iloc[0,1]} emprestimos.')
```
